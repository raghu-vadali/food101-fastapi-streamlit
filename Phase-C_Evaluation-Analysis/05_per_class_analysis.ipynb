```python
import numpy as np
import json

y_true = np.load("y_true.npy")
y_pred = np.load("y_pred.npy")

with open("class_names.json") as f:
    class_names = json.load(f)
```

```python
from collections import defaultdict

class_correct = defaultdict(int)
class_total = defaultdict(int)

for yt, yp in zip(y_true, y_pred):
    class_total[yt] += 1
    if yt == yp:
        class_correct[yt] += 1
        
class_accuracy = {
    class_names[i]: class_correct[i] / class_total[i]
    for i in class_total
}
```

```python
# .items() returns list of tuples (class_name, accuracy) and take accuracy for sorting
sorted_classes = sorted(
    class_accuracy.items(), key=lambda x: x[1]
)

worst_10 = sorted_classes[:10]
best_10 = sorted_classes[-10:]
```

The model performs strongly on visually distinct dishes such as waffles and sushi, while performance drops on visually similar categories such as chicken curry and chicken tikka masala, indicating fine-grained texture confusion.

This per-class analysis highlights systematic performance differences and guides targeted error analysis in subsequent steps.
