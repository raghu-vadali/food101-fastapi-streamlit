Goal of this step

When the model is wrong, what does it confuse the class with?

```python
# Loading cached data
import numpy as np
import json
from collections import defaultdict

y_true = np.load("y_true.npy")
y_pred = np.load("y_pred.npy")

with open("class_names.json") as f:
    class_names = json.load(f)
```

```python
# Building confusion pairs (core logic)
confusion_counts = defaultdict(int)

for yt, yp in zip(y_true, y_pred):
    if yt != yp:
        confusion_counts[(yt, yp)] += 1    # confusion pair: (true class, predicted class)
```

```python
# Sorting most common confusions
sorted_confusions = sorted(
    confusion_counts.items(), key=lambda x: x[1], reverse=True
)

sorted_confusions
```

```python
# Convert to human-readable pairs
confusions = [
    (class_names[t], class_names[p], count)
    for (t, p), count in sorted_confusions
]
top_confusions = confusions[:10]
top_confusions
```

A count only tells you how many mistakes happened, but not how bad is the model relative to the size or structure of the data.

```python
# Normalize confusion by true class frequency
true_class_counts = defaultdict(int)
for yt in y_true:
    true_class_counts[yt] += 1

normalized_confusions = [
    (class_names[t], class_names[p], count / true_class_counts[t])
    for (t, p), count in sorted_confusions
]

normalized_confusions[:10]
```

Confusion pair analysis identifies systematic misclassification patterns by examining which classes are most frequently predicted in place of the true label.
